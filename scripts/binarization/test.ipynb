```python
# Re-import necessary libraries since the environment was reset
import cv2
from matplotlib import pyplot as plt

# Path to the new image
new_image_path = 'image.png'

# Read the new image
new_image = cv2.imread(new_image_path, cv2.IMREAD_GRAYSCALE)

# Binarize the image using a threshold
_, binary_new_image = cv2.threshold(new_image, 127, 255, cv2.THRESH_BINARY)

# Detect contours in the new binarized image
contours_new, _ = cv2.findContours(binary_new_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on a copy of the new image (in color)
contour_new_image = cv2.cvtColor(new_image, cv2.COLOR_GRAY2BGR)
cv2.drawContours(contour_new_image, contours_new, -1, (0, 255, 0), 2)

# Display the image with contours
plt.figure(figsize=(8, 8))
plt.imshow(contour_new_image)
plt.title('Contours Detected in New Image')
plt.axis('off')
plt.show()
```

```python
# Create a blank mask for filling contours
filled_contours_new = cv2.cvtColor(new_image, cv2.COLOR_GRAY2BGR)
cv2.drawContours(filled_contours_new, contours_new, -1, (255, 255, 255), thickness=cv2.FILLED)

# Convert to grayscale and binarize
filled_contours_gray_new = cv2.cvtColor(filled_contours_new, cv2.COLOR_BGR2GRAY)
_, filled_binary_new_image = cv2.threshold(filled_contours_gray_new, 127, 255, cv2.THRESH_BINARY)

# Display the filled and binarized image
plt.figure(figsize=(8, 8))
plt.imshow(filled_binary_new_image, cmap='gray')
plt.title('Filled Contours and Binarized (New Image)')
plt.axis('off')
plt.show()
```
